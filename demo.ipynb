```python
import os
import pickle
import matplotlib.pyplot as plot
from sklearn import neighbors
import numpy
```

```python
%reload_ext autoreload
%autoreload 2
import demo
```

```python
# KNN

face_model_path = 'face_knn_model.clf'
iris_model_L_path = 'iris_knn_model_L.clf'
iris_model_R_path = 'iris_knn_model_R.clf'
```

```python
face_model: neighbors.KNeighborsClassifier = pickle.load(open(os.path.join(face_model_path), 'rb'))
iris_model_L: neighbors.KNeighborsClassifier = pickle.load(open(os.path.join(iris_model_L_path), 'rb'))
iris_model_R: neighbors.KNeighborsClassifier = pickle.load(open(os.path.join(iris_model_R_path), 'rb'))
```

```python
subject_entity = 'S0012_M_54'
```

```python
face, face_feature, iris_L, iris_feature_L, iris_R, iris_feature_R = demo.get_biometrics(
    subject_entity)

plot.figure()
plot.subplot(1, 3, 1)
plot.imshow(face)
plot.subplot(1, 3, 2)
plot.imshow(iris_L)
plot.subplot(1, 3, 3)
plot.imshow(iris_R)

face_prediction = face_model.predict([face_feature])
iris_L_prediction = iris_model_L.predict([iris_feature_L])
iris_R_prediction = iris_model_R.predict([iris_feature_R])

print('Match:')
print('Face:', bool(numpy.sum(face_prediction == subject_entity)))
print('Iris_L:', bool(numpy.sum(iris_L_prediction == subject_entity)))
print('Iris_R:', bool(numpy.sum(iris_R_prediction == subject_entity)))
```
